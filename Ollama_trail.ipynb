```python
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.llms import Ollama
from IPython.display import Markdown

template = """
{messages}
"""

prompt = ChatPromptTemplate.from_messages([("system":"provide answer to the questions"), ("user": "{question}")),

model = Ollama(model="llama3.2")

chain = prompt | model

final_prompt = prompt.format(question="What is happiness?")  # Format the prompt
print(final_prompt)  # Print to check the prompt before sending

result = chain.invoke({"question": "What is happiness?"})
display(Markdown(result))



"""
display(Markdown(chain.invoke({"question":"What is happiness"})))
"""
```
